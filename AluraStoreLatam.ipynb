## 1. Análisis de facturación



En este caso se calcula la facturación total para cada una de las cuatro tiendas con el fin de determinar cuales fueron las tiendas con mayor facturación. En la gráfica se incluye el promedio de facturación de las cuatro tiendas para observar con facilidad que tiendas estuvieron por debajo del promedio de facturación y cuáles por encima.

```python
facturacion = []
for i, df_tienda_i in enumerate(tiendas):
    fact_tienda_i = sum(df_tienda_i['Precio'])
    facturacion.append(fact_tienda_i)
    print(f"Facturación del la tienda {i+1}: ${fact_tienda_i:0.0f}")
```

```python
# Convertir escala a millones para mayor entendimiento
fact_mill = [val/1e6 for val in facturacion]

# Promedio de envio de todas las tiendas
prom_fact = sum(fact_mill)/len(fact_mill)

# Gráfico que resume facturación por tienda
plt.figure(figsize=(12,4))
plt.barh(name_tiendas, fact_mill, color='skyblue')
# Líneas horizontales
plt.axvline(prom_fact, color='red', linestyle='--', linewidth=1, label=f'Promedio: {prom_fact}')
plt.title("Facturación por tienda")
plt.xlabel("Facturación [Millones]")
plt.ylabel("Tienda")
plt.grid(axis='x')
plt.show()
```

En este caso se observa que, de mayor a menor, las tiendas con mayor facturación son:

- Tienda 1
- Tienda 2
- Tienda 3
- Tienda 4

## 2. Ventas por categoría

A continuación, se calcula la cantidad de ventas de productos por cada categoría, luego se lleva a cabo un gráfico de pastel para observar la proporción de ventas de cada una de las categorías.

```python
# Función de cantidad de ventas por categoría 
def cantidad_ventas_x_categoria(tienda_x:pd.DataFrame) ->list:
    """Función que entrega una lista de tuplas iterables con la combinación de (Categoría de producto, cantidad de ventas)

        Argumentos: Dataframe de la tienda
        Return: Lista con las combinaciones de (Categoría de producto, cantidad de ventas)
    """
    cat_productos = sorted(list(tienda_x['Categoría del Producto'].unique()))
    
    cantidad_vendida = []
    for categoria in cat_productos:
        # Filtro para extraer los registros que pertenecen a la categoría
        elementos_categoria = tienda_x[tienda_x["Categoría del Producto"]== categoria]
        cantidad_vendida.append(len(elementos_categoria))
        
    return list(zip(cat_productos, cantidad_vendida))


# Recorrido por cada tienda
for i, tienda_i in enumerate(tiendas):
    print(f"Tienda {i+1}:\n")
    # Inicialización de variables para indicar categoría más popular
    cat_mas_popular = ''
    mayor_cantidad_ventas = 0
    
    #Ventas por categoria
    ventas_por_cat = cantidad_ventas_x_categoria(tienda_i)
    #Recorrido para mostrar ventas
    for info_cat in ventas_por_cat:
        print(f" - Categoría: {info_cat[0]} - Cantidad ventas: {info_cat[1]}")
        if info_cat[1] > mayor_cantidad_ventas:
            cat_mas_popular, mayor_cantidad_ventas = info_cat
            
    #Categoría más popular de la tienda
    print(f"\nLa categoría más popular de la tienda es {cat_mas_popular.upper()} con {mayor_cantidad_ventas} ventas")
    
    #Gráfica de torta para mostrar la proporción de venta por categoría
    plt.figure(figsize=(8,8))
    categorias, valores = list(zip(*ventas_por_cat))
    plt.pie(valores, labels=categorias, autopct="%1.1f%%", startangle=90)
    plt.title("Distribución por categoría")
    plt.axis("equal")
    plt.show()
    
    # Dividir cada tienda con una línea
    print("-"*110 + "\n")   
```

En este caso se observa que la categoría más popular en todas las tiendas fue la categoría de **muebles**.

## 4. Productos más y menos vendidos

En este ejercicio se grafica la cantidad vendida de productos por tienda y además se ilustran los más y menos populares.

```python
# Función para obtener cantidad de ventas por producto
def cantidad_ventas_x_producto(productos:list) -> dict:
    """Función que calcula la cantidad de ventas por cada producto de la tienda
    
    Arguments: 
        productos:list -> Lista con el registro de productos vendidos en la tienda
    
    Return: Diccionario donde la clave es el producto y el valor es la cantidad de ventas
    """
    cantidad_ventas = {}
    for producto in productos:
        cantidad_ventas[producto] = cantidad_ventas.get(producto,0) + 1
        
    return cantidad_ventas

# Función para obtener los más y menos populares
def mas_y_menos_vendidos(cant_ventas:dict) -> dict:
    """Función que retorna cuales fueron los productos que más y menos fueron vendidos
    
    Arguments: 
        cant_ventas:dict -> Diccionario con los productos vendidos por la tienda
    
    return: diccionario con la información del más y el menos vendido"""
    
    # Inicialización de cantidad de vendidos
    dict_vendidos = {'mas_vendido': [('',0)],
                     'menos_vendido': [('', 1e6)]
                     }
    
    # Recorrido de cada uno de los productos
    for producto, ventas in cant_ventas.items():
        # Si supera al actual de más vendidos
        if ventas > dict_vendidos['mas_vendido'][0][1]:
            dict_vendidos['mas_vendido'] = [(producto, ventas)]
        # Si supera al actual de menos vendidos
        elif ventas < dict_vendidos['menos_vendido'][0][1]:
            dict_vendidos['menos_vendido'] = [(producto, ventas)]
        # Si tiene la misma cantidad que el actual de más vendidos
        elif ventas == dict_vendidos['mas_vendido'][0][1]:
            dict_vendidos['mas_vendido'].append((producto, ventas))
        # Si tiene la misma cantidad que el actual de menos vendidos
        elif ventas == dict_vendidos['menos_vendido'][0][1]:
            dict_vendidos['menos_vendido'].append((producto, ventas))
            
    return dict_vendidos
```

```python
# Recorrido para ilustrar la información por tienda
for i, tienda in enumerate(tiendas):
    # Cantidad de ventas por producto  
    ventas_por_producto_i = cantidad_ventas_x_producto(list(tienda['Producto']))
    
    # mas y menos vendidos
    resumen_ventas = mas_y_menos_vendidos(ventas_por_producto_i)
    valor_max = resumen_ventas['mas_vendido'][0][1]
    valor_min = resumen_ventas['menos_vendido'][0][1]
    print(f"RESUMEN TIENDA {i+1}")
    
    # Gráfico de barras
    plt.figure(figsize=(16,6))
    plt.bar(ventas_por_producto_i.keys(), ventas_por_producto_i.values())
    # Líneas horizontales
    plt.axhline(valor_max, color='green', linestyle='--', linewidth=1, label=f'Máximo: {valor_max}')
    plt.axhline(valor_min, color='red', linestyle='--', linewidth=1, label=f'Mínimo: {valor_min}')
    plt.title(f'Cantidad de ventas por Producto Tienda {i+1}')
    plt.xlabel('Producto')
    plt.ylabel('Cantidad Vendida')
    plt.xticks(rotation=90)
    plt.show()
    
    # Impresión de los valores para el resumen
    print(f"Mas vendidos:")
    for info_producto in resumen_ventas['mas_vendido']:
        print(f"- Producto: {info_producto[0]} - Cantidad: {info_producto[1]}")
    print(f"Menos vendidos:")
    for info_producto in resumen_ventas['menos_vendido']:
        print(f"- Producto: {info_producto[0]} - Cantidad: {info_producto[1]}")
        
    print("-"*170 + '\n'*2)
```

## 5. Envío promedio por tienda

En este punto se calcula el costo de envío promedio por cada tienda y se elabora un gráfico de barras para poder comparar visualmente con el promedio de todas las tiendas. 

```python
# Recorrido para calcular e ilustar los datos promedio por tienda
prom_envios = []
for i, df_tienda_i in enumerate(tiendas):
    envio_prom = sum(df_tienda_i['Costo de envío'])/len(df_tienda_i)
    prom_envios.append(envio_prom)
    print(f"Costo de envío promedio del la tienda {i+1}: ${envio_prom:.0f}")

# Promedio de envio de todas las tiendas
prom_envios_all = sum(prom_envios)/len(prom_envios)

# Gráfico de barras
plt.figure(figsize=(8,6))
plt.bar(name_tiendas, prom_envios)
# Líneas horizontales
plt.axhline(prom_envios_all, color='red', linestyle='--', linewidth=1, label=f'Promedio: {prom_envios_all}')
plt.title(f'Promedio de valor de envío por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Valor envío')
plt.show()
```

En este caso las tiendas con mayor y menor costo promedio de envío fueron la **tienda 1** y **tienda 4**, respectivamente.
